```python
%pylab inline
import numpy as np
import scipy.stats as stats
```

```python
def lcm(a, c, m, x_0):
    x_i = x_0
    while True:
        x_i = (a * x_i + c) % m
        yield x_i / m
```

```python
generator = lcm(2, 3, 10, 5)
```

```python
arr = []
generator = lcm(17, 43, 100, 27)
for i in range(0, 1000):
    num = next(generator)
    arr.append(num)
arr = np.array(arr)
```
